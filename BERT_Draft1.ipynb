```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Sample dataset (replace this with your actual dataset)
# data = {
#     'userName': ['143saf', 'AB25', 'abhijitgang', 'agteakash', 'alankarMIS'],
#     'major': ['Robotics', 'MIS', 'MIS', 'MIS', 'MIS'],
#     'researchExp': [0, 0, 0, 0, 0],
#     'industryExp': [18, 66, 0, 0, 0],
#     'internExp': [5.0, 0.0, 0.0, 0.0, 0.0],
#     'greV': [160.0, 146.0, 89.25, 150.0, 147.0],
#     'greQ': [167.0, 157.0, 163.625, 161.0, 156.0],
#     'journalPubs': [0, 0, 0, 0, 0],
#     'confPubs': [0, 0, 0, 0, 0],
#     'cgpa': [0.85, 0.7828, 0.57, 0.622, 0.52],
#     'univName': ['MIT', 'Stanford', 'Harvard', 'Caltech', 'Carnegie Mellon'],
#     'admit': [1, 1, 1, 1, 1]
# }

data = pd.read_csv('user_user_wo_toefl_score.csv')
df = pd.DataFrame(data)
print('data loaded')

# Function to generate BERT embeddings for textual data
def generate_bert_embeddings(texts):
    return model.encode(texts)

# Combine textual columns into a single string for each row
text_columns = ['major', 'researchExp', 'industryExp', 'internExp', 'greV', 'greQ', 'journalPubs', 'confPubs', 'cgpa']
df['combined_text'] = df[text_columns].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

# Split dataset into train and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Reset indices for train and test DataFrames
train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)

# Calculate BERT embeddings for train and test sets
train_embeddings = generate_bert_embeddings(train_df['combined_text'])
test_embeddings = generate_bert_embeddings(test_df['combined_text'])

# Calculate similarity matrix between train and test embeddings
similarity_matrix = cosine_similarity(test_embeddings, train_embeddings)

# Recommend top 5 unique universities for each applicant in the test set
recommendations = {}
confidence_scores = {}

for i, user in enumerate(test_df['userName']):
    similar_users_indices = np.argsort(similarity_matrix[i])[::-1][0:150]  # Get top 150
    similar_users_admissions = train_df.iloc[similar_users_indices]['univName'].values
    unique_universities = np.unique(similar_users_admissions)[:5]  # Get unique top 5
    recommendations[user] = unique_universities
    confidence_scores[user] = cosine_similarity(test_embeddings[i].reshape(1, -1), train_embeddings[similar_users_indices]).flatten()

# Get usernames and corresponding university vectors
user_univ_vectors = {}
for user, univs in recommendations.items():
    user_univ_vectors[user] = train_df[train_df['univName'].isin(univs)][['userName', 'univName']].values

# Print the usernames, their related university vectors, and the confidence scores
for user, univs in user_univ_vectors.items():
    print(f"User: {user}")
    unique_univs, unique_scores = np.unique(univs[:, 1], return_index=True)
    for u, score in zip(unique_univs, unique_scores):
        score_index = np.where(recommendations[user] == u)[0][0]
        print(f"Related University: {u}, Confidence Score: {confidence_scores[user][score_index]:.4f}")

# Calculate RMSE
actual_universities = test_df['univName'].values
predicted_universities = []

for user, univs in recommendations.items():
    if test_df[test_df['userName'] == user]['univName'].values[0] not in univs:
        univs = np.append(univs, test_df[test_df['userName'] == user]['univName'].values[0])
    predicted_universities.append(univs)

```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

predicted_labels = []
for i in range(len(test_df)):
    # Check if i is within the range of predicted_universities
    if i < len(predicted_universities):
        for univ in predicted_universities[i]:
            if univ == test_df.iloc[i]['univName']:
                predicted_labels.append(1)
                break  # Exit the loop once a match is found
        else:
            predicted_labels.append(0)  # Append 0 if no match is found
    else:
        predicted_labels.append(0)  # Append 0 for out-of-range indices

actual_labels = np.ones(len(predicted_labels))

rmse = np.sqrt(mean_squared_error(actual_labels, predicted_labels))
print(f"RMSE: {rmse:.4f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convert predicted_labels and actual_labels to numpy arrays
predicted_labels_np = np.array(predicted_labels)
actual_labels_np = np.array(actual_labels)

# Calculate accuracy
accuracy = accuracy_score(actual_labels_np, predicted_labels_np)

# Calculate precision
precision = precision_score(actual_labels_np, predicted_labels_np)

# Calculate recall
recall = recall_score(actual_labels_np, predicted_labels_np)

# Calculate F1 score
f1 = f1_score(actual_labels_np, predicted_labels_np)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
